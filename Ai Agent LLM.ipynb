```python
import smtplib
from datetime import datetime
from email.mime.text import MIMEText
from openai import OpenAI
```

```python
team_birthdays = {
    "Nita": ("nita@example.com", "1998-03-29"),
    "Alice": ("alice@example.com", "1990-06-15"),
    "Bob": ("bob@example.com", "1988-11-02"),
    "Charlie": ("charlie@example.com", "1992-09-10"),
    "David": ("david@example.com", "1995-12-25")
}
```

```python
client = OpenAI(api_key="OPENAI_API_KEY")
```

```python
def generate_birthday_message(name):
    prompt = f"Write a warm, friendly, and personal birthday wish for {name}."
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
    )
    return response.choices[0].message.content
```

```python
def send_email(to_email, subject, message):
    sender_email = "your_email@gmail.com"
    sender_password = "your_app_password"  # Use app password!

    msg = MIMEText(message)
    msg["Subject"] = subject
    msg["From"] = sender_email
    msg["To"] = to_email

    with smtplib.SMTP_SSL("smtp.gmail.com", 465) as server:
        server.login(sender_email, sender_password)
        server.sendmail(sender_email, to_email, msg.as_string())
    print(f"Email sent to {to_email}")
```

```python
def check_and_send_birthday_wishes():
    today = datetime.now().strftime("%m-%d")

    for name, (email, birthday) in team_birthdays.items():
        bday_md = datetime.strptime(birthday, "%Y-%m-%d").strftime("%m-%d")
        if bday_md == today:
            message = generate_birthday_message(name)
            subject = f"Happy Birthday {name}! 🎉"
            send_email(email, subject, message)
```

```python
if __name__ == "__main__":
    check_and_send_birthday_wishes()
```
